```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
cars_data = pd.read_csv('USA_cars_datasets.csv')
cars_data.head(20)
#printing the first 20 rows
```

```python
carYear = cars_data['year']
```

```python
# subset of car sales from 2018-2020
cars_in_2018 = cars_data[cars_data['year']==2018].set_index('model')
cars_in_2019 = cars_data[cars_data['year']==2019].set_index('model')
cars_in_2020 = cars_data[cars_data['year']==2020].set_index('model')


car_by_year_2018 = cars_in_2018[['brand','state', 'year','mileage','price']]
car_by_year_2019 = cars_in_2019[['brand','state', 'year','mileage','price']]
car_by_year_2020 = cars_in_2020[['brand','state', 'year','mileage','price']]

car_by_year_2020.head(20)
```

```python
#What car brand was sold the most in 2020?
car_by_year_2020['brand'].value_counts()
```

```python
#Bar chart showing car brands sold in 2020 using the subset
car_by_year_2020['brand'].value_counts().plot(kind='bar', figsize=(7, 6), rot=0)
plt.xlabel("Brands", labelpad=14)
plt.ylabel("Count of car brands", labelpad=14)
plt.title("Car brands sold in 2020", y=1.02);
```

```python
#Bar chart showing car brands sold in 2019 using the subset
car_by_year_2019['brand'].value_counts().plot(kind='bar', figsize=(15, 6), rot=0)
plt.xlabel("Brands", labelpad=14)
plt.ylabel("Count of car brands", labelpad=14)
plt.title("ar brands sold in 2019", y=1.02);
```

```python
car_by_year_2018['brand'].value_counts().plot(kind='bar', figsize=(15, 6), rot=0)
plt.xlabel("Brands", labelpad=14)
plt.ylabel("Count of car brands", labelpad=14)
plt.title("Car brands sold in 2018", y=1.02);
```
